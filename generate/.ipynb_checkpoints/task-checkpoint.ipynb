```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('about.csv')
display(data)
```

```python
def get_link(url):
    url = url.strip()
    return f'[{url}]({url})'
```

```python
for index, row in data.iterrows():
    block = '---\n'
    block += '<div style="display: flex;" markdown>\n'
    block += '<img src="/people_thumbnail/' \
        + row['thumbnail'] + '" width="200" />\n'
    block += '<div style="width: 5%">\n'
    block += '</div>\n'
    
    block += '<div markdown style="width: 70%;">\n'
    block += '**Name**: ' + row['name'] + '\n\n'
    
    block += '**Title**: ' + row['title'] + '\n\n'
    if row['linkedin'] is not np.nan:
        block += '**LinkedIn**: ' + get_link(row['linkedin']) + '\n\n'
      
    if row['website'] is not np.nan:
        block += '**Personal Website**: ' + get_link(row['website']) + '\n\n'
    
    if row['twitter'] is not np.nan:
        block += '**Twitter**: ' + get_link(row['twitter']) + '\n\n'
        
    if row['facebook'] is not np.nan:
        block += '**Facebook**: ' + get_link(row['facebook']) + '\n\n'
        
    if row['section'] is in ['6.1.1.1', '6.1.1.2']
        block += '!!! quote "Motto in light of eastern and western philosophy"\n\n'
    
    block += '</div>\n</div>\n'
    print(block)
```
